```python
def money_mentor():
    currencies = {
        "INR": 1.00,
        "USD": 93.84,
        "EUR": 108.74,
        "GBP": 125.70,
        "JPY": 0.59,
        "CNY": 13.64,
        "CAD": 68.26,
        "AUD": 65.40,
        "SGD": 73.36,
        "AED": 25.55,
        "SAR": 25.00,
        "ZAR": 5.52,
        "RUB": 1.16,
        "BRL": 17.83,
        "MXN": 5.27,
        "KRW": 0.025
    }

    # currency input
    currency = input("Enter currency (INR/USD/EUR...): ").upper()

    if currency not in currencies:
        print("Currency not supported")
        return

    income = float(input("Enter salary: "))
    expenses = float(input("Enter expenses: "))
    savings = float(input("Enter savings: "))

    # convert to INR
    rate = currencies[currency]
    income_inr = income * rate
    expenses_inr = expenses * rate
    savings_inr = savings * rate

    # calculations
    savings_percent = (savings_inr / income_inr) * 100
    expense_percent = (expenses_inr / income_inr) * 100

    # scoring
    if savings_percent < 20:
        score = "Poor"
    elif savings_percent < 30:
        score = "Good"
    else:
        score = "Excellent"

    # output
    print(f"\nCurrency Used: {currency}")
    print(f"Income: {income} {currency}")
    print(f"Expenses: {expenses} {currency}")
    print(f"Savings: {savings} {currency}")
    print(f"Income in INR: ₹{income_inr:.2f}")
    print(f"Savings Percentage: {savings_percent:.2f}%")
    print(f"Expense Percentage: {expense_percent:.2f}%")
    print(f"Financial Health: {score}")

    # suggestions
    if savings_percent < 20:
        print("- Try to reduce unnecessary expenses")
        print("- Aim to save at least 20% of income")
    elif savings_percent < 30:
        print("- You are doing good, try to increase savings slightly")
    else:
        print("- Excellent savings habit! Consider investing")

    # warning
    if expense_percent > 70:
        print("⚠ Warning: Expenses are too high!")
```

```python
money_mentor()
```
